### We analyse the transactions data from a shopping mall and try to find out association rules using apriori algorithm. 

```python
# Import Libraries

from apyori import apriori
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data

data=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
```

```python
data.head()
```

```python
data.values
```

```python
#Make the list of lists of transactions

transactions=[]
for i in range(len(data)):
    reclist=[str(data.values[i,j]) for j in range(len(data.columns)) if str(data.values[i,j])!='nan']
    transactions.append(reclist)
transactions[0:5]
        
        
```

```python
# Find rules using apriori algorithm

rules=list(apriori(transactions,min_support=0.003,min_confidence=0.20,min_lift=3))
```

```python
rules
```

```python
# Create dataframe out of rules obtained

itembase=[rules[i][2][0][0] for i in range(len(rules))]
itemadded=[rules[i][2][0][1] for i in range(len(rules))]    
df=pd.DataFrame()
df['item base']=itembase
df['item added']=itemadded
```
