```python
import pandas as pd
data=pd.DataFrame(pd.read_csv('credit_train.csv'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data.isnull().sum()
```

```python
# Since Loaan Id and Customer Id are irrelevant to response, let us drop them

data.drop(['Loan ID','Customer ID'],axis=1,inplace=True)
data.info()
```

```python
# Drop rows where response variable (Loan Status ) is null

import numpy as np
#data=data[np.isfinite(data['Loan Status'])]
data.dropna(subset=['Loan Status'],inplace=True)
```

```python
data.shape
```

```python
data.info()
```

```python
data['Annual Income'].hist(bins=100)
```

```python
# Annual income has skewed distribution. so let us fill it by median

data['Annual Income']=data['Annual Income'].fillna(data['Annual Income'].median())
```

```python
data.info()
```

```python
data['Number of Credit Problems'].corr(data['Bankruptcies'])
```

```python
# Since 'Number of Credit Problems and 'Bankruptcies' are correlated , we drop one of them . Let us drop 'Number of Credit Problems'
data.drop('Number of Credit Problems',axis=1,inplace=True)
```

```python
data.info()
```

```python
# Let us convert 'Years in current job' column to numeric. Also Replace '<' value by 'nan'

data['Years in current job']=data['Years in current job'].apply(lambda x:str(x)[0:2])
#data.head()
data['Years in current job']=data['Years in current job'].apply(lambda x:str.replace(x,'<','nan'))
data['Years in current job']=pd.to_numeric(data['Years in current job'],errors='coerce')
```

```python
#data['Years in current job']
```

```python
data.info()
```

```python
data['Years in current job'].hist(bins=10)
```

```python

data['Years in current job']=data['Years in current job'].fillna(data['Years in current job'].median())
```

```python
data.info()
```

```python
# Let us explore distributions of features 

import matplotlib

plt.subplot(2,3,1)
plt.title('Credit Score')
plt.hist(data['Credit Score'],bins=100)

plt.subplot(2,3,2)
plt.title('Months since last delinquent')
plt.hist(data['Months since last delinquent'],bins=50)

plt.subplot(2,3,3)
plt.title('Maximum Open Credit')
plt.hist(data['Maximum Open Credit'],bins=50)

plt.subplot(2,3,4)
plt.title('Bankruptcies')
plt.hist(data['Bankruptcies'],bins=50)

plt.subplot(2,3,5)
plt.title('Tax Liens')
plt.hist(data['Tax Liens'],bins=100)
plt.subplots_adjust(wspace=1,hspace=1)
```

```python
data['Credit Score']=data['Credit Score'].fillna(data['Credit Score'].median())
data['Months since last delinquent']=data['Months since last delinquent'].fillna(data['Months since last delinquent'].median())
data['Maximum Open Credit']=data['Maximum Open Credit'].fillna(data['Maximum Open Credit'].median())
data['Bankruptcies']=data['Bankruptcies'].fillna(data['Bankruptcies'].mode()[0])
```

```python
data.info()
```

```python
data['Tax Liens'].value_counts()
```

```python
# Tax Liens does not contribute much to the response. So let us drop it

data.drop('Tax Liens',axis=1,inplace=True)
```

```python
data.info()
```

## Since the missing values are handled, let us encode categorical features

```python
cat_cols=['Loan Status','Term','Home Ownership','Purpose']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat_cols:
    data[i]=le.fit_transform(data[i])
    
```

```python
data.head()
```

### Normalise Data 
### Apply minmax scaler since data is skewed



```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mmscaler=MinMaxScaler()
```

```python
X=data.drop('Loan Status',axis=1)
y=data['Loan Status']
scaled_data=mmscaler.fit_transform(X)
X.columns
```

```python
scaled_data
```

```python
# Let us split data into training and testing sets

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(scaled_data,y,test_size=0.3)
```

```python
# Build classifiers and train and evaluate them

from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
adb=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=50,learning_rate=0.01,random_state=100)
```

```python
adb.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
ypred=adb.predict(xtest)
adb.score(xtest,ytest)
```

```python
#  Let us plot Feature importance 
fimp=pd.concat([pd.Series(data.columns),pd.Series(adb.feature_importances_)],axis=1)
fimp.columns=['feature','imp score']
#fimp
fimp.plot(kind='bar')
```

## Now it's time for tuning hyperparameters. We tune parameters for three classifiers

1. AdaBoostClassifier
2. RandomForestClassifier
3. SVC


```python
from sklearn.model_selection import GridSearchCV

param_grid_adb = {
 'learning_rate':[0.1,0.01,0.001,0.0001,0.00001]
}
clf1 = GridSearchCV(estimator = adb,param_grid = param_grid_adb, scoring='roc_auc',n_jobs=-1, cv=5,refit=True)
clf1.fit(xtrain,ytrain)
print(clf1.best_estimator_, clf1.best_params_, clf1.best_score_)
clf1.score(xtest,ytest)
```

```python
# So we obtained optimal learning_rate=0.1
```

##   RandomForestClassifier

Let us tune following parameters

1. max_features
2. min_samples_leaf


```python
# param_grid_rfc={'max_features':range(5,8),'min_samples_leaf':range(10,50,10)}
```

```python
# Let us tune max_features

param_grid_rfc_feat={'max_features':[5,6,7,8]}
rfc=RandomForestClassifier(random_state=100)
clf2 = GridSearchCV(estimator=rfc, param_grid=param_grid_rfc_feat,scoring='roc_auc',n_jobs=-1,cv=5,refit=True)
clf2.fit(xtrain,ytrain)
```

```python
clf2.best_estimator_,clf2.best_params_,clf2.best_score_
```

```python
# Let us tune min_samples_leaf by keeping max features=7

rfc=RandomForestClassifier(max_features=7,random_state=100)
param_grid_rfc_samples={'min_samples_leaf':[10,20,30,40,50]}
clf2 = GridSearchCV(estimator=rfc, param_grid=param_grid_rfc_samples,scoring='roc_auc',n_jobs=-1,cv=5,refit=True)
clf2.fit(xtrain,ytrain)
```

```python
clf2.best_estimator_,clf2.best_params_,clf2.best_score_
```

# So we obtained optimal values for hyperparameters as 

max_features=7
min_samples_leaf=50
